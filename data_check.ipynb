```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# 1. Load the Dataset
data_path = '/root/timesnet/Time-Series-Library/data/new_stock_data.csv'
df = pd.read_csv(data_path, parse_dates=['date'])
df.set_index('date', inplace=True)
print(df.head())
print(f"Total number of rows in the dataset: {len(df)}")  # Should be 1832

# 2. Load and Aggregate Predictions
folder_path ='/root/timesnet/Time-Series-Library/test_results/long_term_forecast_test_TimesNet_custom_ftMS_sl60_ll30_pl7_dm512_nh8_el2_dl1_df1024_expand2_dc4_fc1_ebtimeF_dtTrue_test_0'
preds = np.load(os.path.join(folder_path, 'pred.npy'))  # Shape: (360, 7, 1)
trues = np.load(os.path.join(folder_path, 'true.npy'))  # Shape: (360,)
print(f"Predictions shape: {preds.shape}")  # Expected: (360, 7, 1)
print(f"True values shape: {trues.shape}")  # Expected: (360,)

# Aggregate predictions
aggregated_preds_last = preds[:, -1, 0]
aggregated_preds_avg = preds.mean(axis=1).squeeze()
print(f"Aggregated Predictions (Last) shape: {aggregated_preds_last.shape}")  # (360,)
print(f"Aggregated Predictions (Average) shape: {aggregated_preds_avg.shape}")  # (360,)

# Choose aggregation method
aggregated_preds = aggregated_preds_last  # or aggregated_preds_avg

# 3. Align Predictions with the Dataset
assert len(df) == 1832, "The dataset does not contain 1832 rows as expected."
assert aggregated_preds.shape[0] == 360, "The number of predictions does not match the test set size."

# Add 'Predictions' column initialized with NaN
df['Predictions'] = np.nan

# Define the test set's starting row index (0-based)
test_start_row = 1466  # Corresponds to row 1468

# Assign predictions to the 'Predictions' column for the test set
df.iloc[test_start_row:test_start_row + 360, df.columns.get_loc('Predictions')] = aggregated_preds

# Optional: Verify the assignment
print(df[['收盘', 'Predictions']].iloc[test_start_row:test_start_row + 5])

# 4. Visualize the Data with English Labels

# 4.1. Plotting the Entire Dataset
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16, 8))
plt.plot(df.index, df['收盘'], label='True Close Price', color='blue')
plt.plot(df.index, df['Predictions'], label='Predicted Close Price', color='red', linestyle='--')
plt.axvline(df.index[test_start_row - 1], color='green', linestyle='--', label='End of Train Set')
plt.axvline(df.index[test_start_row + 366 - 1], color='orange', linestyle='--', label='End of Test Set')
plt.legend()
plt.title('Stock Close Price vs. Predictions')
plt.xlabel('Date')
plt.ylabel('Price')
plt.gcf().autofmt_xdate()
plt.grid(True)
plt.tight_layout()
plt.show()

# 4.2. Plotting the Test Set in Detail
test_start_date = df.index[test_start_row]
test_end_date = df.index[test_start_row + 366 - 1]
mask = (df.index >= test_start_date) & (df.index <= test_end_date)
plt.figure(figsize=(16, 8))
plt.plot(df.loc[mask].index, df.loc[mask, '收盘'], label='True Close Price', color='blue')
plt.plot(df.loc[mask].index, df.loc[mask, 'Predictions'], label='Predicted Close Price', color='red', linestyle='--')
plt.legend()
plt.title('Test Set: Stock Close Price vs. Predictions')
plt.xlabel('Date')
plt.ylabel('Price')
plt.gcf().autofmt_xdate()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 持久模型预测（使用上一期的收盘价作为预测）
persistent_preds = df['收盘'].shift(1).iloc[test_start_row:test_start_row + 360].values

aggregated_trues = trues[:,-1,0]
# 计算持久模型的指标
persistent_mae = mean_absolute_error(aggregated_trues, persistent_preds)
persistent_mse = mean_squared_error(aggregated_trues, persistent_preds)
persistent_rmse = np.sqrt(persistent_mse)
persistent_mape = np.mean(np.abs((aggregated_trues - persistent_preds) / trues)) * 100
persistent_mspe = np.mean(((aggregated_trues - persistent_preds) / trues) ** 2)

print(f"Persistent Model MAE: {persistent_mae}")
print(f"Persistent Model MSE: {persistent_mse}")
print(f"Persistent Model RMSE: {persistent_rmse}")
print(f"Persistent Model MAPE: {persistent_mape}%")
print(f"Persistent Model MSPE: {persistent_mspe}")
```
